```python
# Program Mengira Bilangan Perkataan Dalam Ayat
# Penggunaan fungsi Split() untuk memisahkan perkataan.
''' Apabila tiada nilai diletak pada fungsi split(),
    ruang kosong dijadikan pembahagi perkataan'''

print("Taip 'x' untuk keluar. ");
ayat = input("Memasukkan satu ayat: ");
if ayat == 'x':
    exit();
else:
    panjangAyat = len(ayat.split());
    print("\nTerdapat",panjangAyat,"perkataan dalam ayat ini. ");
```
